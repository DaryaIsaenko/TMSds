```python
# 1. Решить уравнения
def equation(a, b, c):
  d = b**2 - 4 * a * c
  x1 = (-b + d**0.5) / (2 * a)
  x2 = (-b - d**0.5) / (2 * a)
  print(f'x1 = {x1}, x2 = {x2}')

# y**2 - 12y + 20 = 0
equation(1, -12, 20)

# z**2 + 17z + 72 = 0
equation(1, 17, 72)

# x**2 - 7x - 44 = 0
equation(1, -7, -44)

# y**2 + 9y + 8 = 0
equation(1, 9, 8)

# b**2 - 2b - 63 = 0
equation(1, -2, -63)
```

```python
# 2. Представить список в обратном порядке
# переворачиваем список используя срез
arr1 = ['green', 'red', 'blue', 'black']
print(arr1[::-1])

# переворачиваем список с помощью функции reverse()
arr2 = [2, 1, 5.2, 123, 4.5]
arr2.reverse()
print(arr2)
```

```python
# 3. Поменять местами первый и последний элементы списка
lst = [1, 2, 3, 4, 5]

def change(lst):
  if len(lst) >= 2:
    lst[0], lst[-1] = lst[-1], lst[0]
    print(lst)
  else:
    print('Список должен иметь минимум 2 элемента.')

change(lst)
```

```python
# 4. Поиск "бесполезного" числа на основании списка
s = [1, 5, 85, 78, 12, 6, 6, 56]

def useless(s):
  max_val = max(s)
  len_s = len(s)
  useless_number = max_val / len_s
  print(useless_number)

useless(s)
```

```python
# 5. Вывести список из строк одинаковой длины
lst = ['крот', 'белка', 'выхухоль']

def all_eq(lst):
  lst_new = []
  max_len_str = 0 # максимальная длина строки
  for i in lst:
    if len(i) > max_len_str:
      max_len_str = len(i)
  for k in lst:
    if len(k) < max_len_str:
      qnt_symbol = max_len_str - len(k) # количество символов необходимых для добавления
      lst_new.append(k + '_' * qnt_symbol)
    else:
      lst_new.append(k)
  print(lst_new)

all_eq(lst)
```

```python
# сортировка
arr = [5, 1, 85, 54, 2]
counter1 = 0
counter2 = 0

for i in range(len(arr) - 1):
  counter1 += 1
  for k in range(len(arr) - 1 - i):
    counter2 += 1
    if arr[k] > arr[k + 1]:
      arr[k], arr[k + 1] = arr[k + 1], arr[k]

print(arr)
print(f'counter1 = {counter1}, counter2 = {counter2}')
```
