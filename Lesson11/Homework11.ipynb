```python
import numpy as np
```

```python
#1
matrixA = np.array([
    [1, 2, 4],
    [7, 5 ,-2],
    [-2, 1, 3]
])

matrixB = np.array([
    [-1, 0, 3],
    [2, -6, 1],
    [-7, 1, 5]
])

matrixC = np.array([
    [-2, 3],
    [4, 0],
    [7, -1]
])

addition = matrixA + matrixB
subtraction = matrixB - matrixA
multiplication = np.dot(matrixA, matrixC)
multiplicationABC = np.dot(np.dot(matrixA, matrixB), matrixC)

print(f'A + B =\n{addition}\n')
print(f'B - A =\n{subtraction}\n')
print(f'A * C =\n{multiplication}\n')
print(f'A * B * C =\n{multiplicationABC}\n')
```

3*. Необходимо реализовать на python (+numpy) алгоритм матричной свертки (3x3) для представленного изображения.

```python
import matplotlib.pyplot as plt
```

```python
def convolution(image, kernel):
  kernel_height, kernel_weight = kernel.shape
  image_height, image_weight, image_channels = image.shape

  convolution_matrix = np.zeros((image_height - kernel_height + 1, image_weight - kernel_weight + 1, image_channels))

  for ch in range(image_channels):
    for h in range(image_height - kernel_height + 1):
      for w in range(image_weight - kernel_weight + 1):
        convolution_matrix[h, w, ch] = np.sum(image[h:h + kernel_height, w:w + kernel_weight, ch] * kernel).clip(0, 255)

  return convolution_matrix.astype(np.uint8)
```

```python
image = plt.imread('image.jpg')
kernel = np.array([
    [-1, -1, -1],
    [-1, 9, -1],
    [-1, -1, -1]
])

filtered_image = convolution(image, kernel)


plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Origin image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(filtered_image)
plt.title('Image with filter')
plt.axis('off')

plt.show()
```
