```python
df['drivewheel'] = df['drivewheel'].replace("4wd", 'wdf')
df['fuelsystem'] = df['fuelsystem'].replace("2bbl", 'tbbi')
df['fuelsystem'] = df['fuelsystem'].replace("1bbl", 'obbl')
df['fuelsystem'] = df['fuelsystem'].replace("4bbl", 'fbbl')

df_dummie = pd.get_dummies(df, columns = ['aspiration','carbody','drivewheel','enginetype','cylindernumber','fuelsystem'])
df_dummie = df_dummie.drop(['CarName','car_ID','brands'], axis=1)
df_dummie.head()
```

```python
categories_new  = ['doornumber', 'enginelocation',
       'boreratio', 'stroke', 'compressionratio',
       'carbody_hardtop', 'carbody_hatchback', 'carbody_sedan',
       'carbody_wagon', 'drivewheel_rwd', 'drivewheel_wdf', 'enginetype_dohcv',
       'enginetype_l', 'enginetype_ohc', 'enginetype_ohcf', 'enginetype_ohcv',
       'enginetype_rotor', 'cylindernumber_five', 'cylindernumber_four',
       'cylindernumber_six', 'cylindernumber_three', 'cylindernumber_twelve',
       'cylindernumber_two','fuelsystem_idi', 'fuelsystem_mfi',
       'fuelsystem_mpfi', 'fuelsystem_obbl', 'fuelsystem_spdi',
       'fuelsystem_spfi', 'fuelsystem_tbbi','fueltype','fuelsystem_fbbl','cylindernumber_eight',
       'enginetype_dohc','drivewheel_fwd','carbody_convertible','aspiration_turbo','aspiration_std']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for feature in categories_new:
    df_dummie[feature] = le.fit_transform(df_dummie[feature])
```

```python
df_dummie['carwlh'] = df_dummie.carwidth + df_dummie.carlength + df_dummie.carheight
df_dummie = df_dummie.drop(['carwidth','carlength','carheight'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X = df_dummie.drop(['price','citympg','highwaympg'],axis = 1 )
y = df_dummie['price']
X_train,X_test,y_train,y_test =  train_test_split(X,y,train_size = 0.80 , random_state = 42)
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from sklearn.ensemble import RandomForestRegressor
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_absolute_error

mmc = MinMaxScaler()

X_train = mmc.fit_transform(X_train)
X_test = mmc.transform(X_test)
```

<font size=4>**Gradient Boosting Regressor**</font>

```python
from sklearn.metrics import mean_squared_error

params = {
    "n_estimators": 1000,
    "max_depth": 8,
    "learning_rate": 0.01,
    "loss": "absolute_error",
}

ensemble = GradientBoostingRegressor(**params)
ensemble.fit(X_train, y_train)
print('score:', ensemble.score(X_test, y_test))

test_score = np.zeros((params["n_estimators"],), dtype=np.float64)
for i, y_pred in enumerate(ensemble.staged_predict(X_test)):
    test_score[i] = mean_absolute_error(y_test, y_pred)

print('mae:', mean_absolute_error(y_test, ensemble.predict(X_test)))

fig = plt.figure(figsize=(6, 6))
plt.title("Error")
plt.plot(
    np.arange(params["n_estimators"]) + 1,
    ensemble.train_score_,
    "b-",
    label="Training Set Deviance",
)
plt.plot(
    np.arange(params["n_estimators"]) + 1, test_score, "r-", label="Test Set Deviance"
)
plt.legend(loc="upper right")
plt.xlabel("Boosting Iterations")
plt.ylabel("Error")
fig.tight_layout()
plt.show()
```

<font size=4>**Random forest**</font>

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D

train_results = []
test_results = []
list_nb_trees = [5, 1200]

for nb_trees in list_nb_trees:
    rf = RandomForestRegressor(n_estimators=1200, max_depth=9)
    rf.fit(X_train, y_train)

    train_results.append(mean_absolute_error(y_train, rf.predict(X_train)))
    test_results.append(mean_absolute_error(y_test, rf.predict(X_test)))

print('score:', rf.score(X_train,y_train))
print('mae:', mean_absolute_error(y_test, rf.predict(X_test)))

line1, = plt.plot(list_nb_trees, train_results, color="r", label="Training Score")
line2, = plt.plot(list_nb_trees, test_results, color="g", label="Testing Score")

plt.title("Error")
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('MAE')
plt.xlabel('RF Iterations')
plt.show()
```

<font size=4>**Nueral network**</font>

```python
input_shape = (X_train.shape[1],)
model4 = Sequential([
    Dense(units=200, activation='relu',
                    input_shape=input_shape),
    Dense(units=100, activation='relu'),
    Dropout(0.38),
    Dense(units=200, activation='relu'),
    Dense(units=1)
])

optimizer=Adam(learning_rate=0.01)
model4.compile(optimizer, loss=['mse'], metrics = ["mae"])

loss=model4.fit(X_train, y_train, epochs = 500, validation_data = (X_test, y_test))
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.xlabel('Epoch')
  plt.ylabel('Error')
  plt.legend()

plot_loss(loss)
```

```python
from sklearn.metrics import r2_score

y_pred2 = model4.predict(X_test)

print("R2 score:",r2_score(y_test, y_pred2))
print('mae', mean_absolute_error(y_test, y_pred2))
```

<font size=4>**Models**

```python
models = pd.DataFrame(columns=['Model', 'Score Test', 'MAE Test'])

m1 = {'Model': 'GradientBoostingRegressor', 'Score Test': ensemble.score(X_test, y_test), 'MAE Test': mean_absolute_error(y_test, ensemble.predict(X_test))}
models = pd.concat([models, pd.DataFrame([m1])], ignore_index=True)
m2 = {'Model': 'RandomForestRegression', 'Score Test': rf.score(X_train,y_train), 'MAE Test': mean_absolute_error(y_test, rf.predict(X_test))}
models = pd.concat([models, pd.DataFrame([m2])], ignore_index=True)
m3 = {'Model': 'NueralNetwork', 'Score Test': r2_score(y_test, y_pred2), 'MAE Test': mean_absolute_error(y_test, y_pred2)}
models = pd.concat([models, pd.DataFrame([m3])], ignore_index=True)

models
```

```python
plt.subplots(figsize=(6,3))

data = models.sort_values(by='Score Test', ascending=False)

ax = sns.barplot(data=data, x=data['Score Test']*100, y='Model')
ax.bar_label(ax.containers[0], fontsize=8)

plt.title('Rank of models by Score')
plt.xlabel('Score')
plt.show()
```

```python
plt.subplots(figsize=(6,3))

data = models.sort_values(by='MAE Test', ascending=True)
ax = sns.barplot(data=data, x=data['MAE Test'], y='Model')
ax.bar_label(ax.containers[0], fontsize=8)

plt.title('Rank of models by MAE')
plt.xlabel('MAE Test')
plt.show()
```
